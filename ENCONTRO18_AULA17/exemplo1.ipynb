```python
# configurações / pesquisar interactive / 
# Jupyter Interactive Window Text Edition Execute Selection

# importar pandas e numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# Obtendo os dados do ISP
try:
    # Endereço dos dados do isp
    ENDERECO_DADOS = 'https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv'

    # Ler os dados
    df_isp = pd.read_csv(ENDERECO_DADOS, sep=';', encoding='iso-8859-1')

    # Ver as primeiras linhas
    display(df_isp.columns)
except ImportError as e:
    print('Erro ao ler os dados:', e)
```

```python
# Delimitando as variáveis
try:
    df_roubos_veiculos = df_isp[['cisp', 'mes', 'ano', 'munic', 'regiao', 'roubo_veiculo']]

    # totalizando os roubos de veículos por municipio, região e mês
    # df_roubos_veiculos = df_roubos_veiculos.groupby(['cisp']).sum(['roubo_veiculo']).reset_index()

    # df_roubos_veiculos_municipio_ordenado = df_roubos_veiculos_municipio.sort_values(by='roubo_veiculo', ascending=False)

    # display(df_roubos_veiculos.head(10))

except ImportError as e:
    print('Erro ao tratar os dados:', e)
```

```python
# calculando os quartis
try:
    array_roubos_veiculos = np.array(df_roubos_veiculos['roubo_veiculo'])

    total = np.sum(array_roubos_veiculos)
    media = np.mean(array_roubos_veiculos)
    mediana = np.median(array_roubos_veiculos)

    # imprimindo os resultados
    print(f'Total de roubos de veículos: {total}')
    print(f'Média de roubos de veículos: {media}')
    print(f'Mediana de roubos de veículos: {mediana}')

    # quartis
    q1 = np.quantile(array_roubos_veiculos, 0.25, method='weibull')
    q3 = np.quantile(array_roubos_veiculos, 0.75, method='weibull')
        
    print(f'Q1: {q1}')
    print(f'Q3: {q3}')
except ImportError as e:
    print('Erro ao calcular os quartis:', e)
```

```python
# Criando uma coluna nova, adicionando identificadores
# Concatena a nova coluna no datafreme
try:
    # criando a nova coluna
    df_roubos_veiculos['flag'] = np.where(df_roubos_veiculos['roubo_veiculo'] <= q1, 'menos',
                                      np.where(df_roubos_veiculos['roubo_veiculo'] >= q3, 'mais', 'mais ou menos'))
    display(df_roubos_veiculos)

    # salvando o dataframe
    df_roubos_veiculos.to_csv('roubos_veiculos.csv', index=False, sep=';', encoding='iso-8859-1')
except ImportError as e:
    print('Erro ao criar a coluna:', e)
```

```python
# imprimindo um boxplot
try:
    plt.boxplot(array_roubos_veiculos, vert=False, showmeans=True, meanline=True)
    plt.title('Boxplot dos roubos de veículos')
    plt.show()
except ImportError as e:
    print('Erro ao plotar o boxplot:', e)
```
